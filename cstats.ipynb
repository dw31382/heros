Data Types:

<b>Categorical</b>
    Can be divided into groups: <i>race, sex</i>.
    
<b>Numerical</b>
    Quantitative, form of numbers, not descriptive strings
    
 - Continuous
    Impossible, abstract nonsense, uncountable values, but "measurable": height, weight, temperature, length.

 - Discrete
    Countable, individual and discinct, integer enumerable: # of customers, # of computers in each department.

<b>Ordinal</b>
    Categorical data with order: <i>socio economic status, age group, educational level</i>.

    By knowing the data type of your data source, you will be able to know what technique to use when analyzing them.


Mean - The average value </br>
Median - The mid point value </br>
Mode - Most common value


```python
import numpy as np

x = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
```

```python
# calculate mean w/o np
print(sum(x) / len(x))

# calculate mean w/ np
print(np.mean(x))

# median
print(np.median(x))

from scipy import stats
# mode
# returns object w/ mode and count
print(stats.mode(x, keepdims=False).mode)
# keepdims=False prevents a weird message
```

<b>Standard deviation</b> is a measure of how much <u>variation</u> there is in a set of data. The higher the standard deviation, the more spread out the data is. The opposite of average would be a measure of how far the data is from the mean. 

    Standard deviation and variance are measures of how far the data is from the mean.

```python
# standard deviation
print(np.std(x))

# variance
print(np.var(x))
```

Formula standard deviation:

    σ = √Σ (x-μ)^2 / N
Formula for variance:

    σ^2 = Σ (x-μ)^2 / N
where,

    x = individual data points
    μ = mean
    Σ = sum
    N = number of data points

A <b>percentile</b> is a measure of where a particular value falls within a set of values.

    Percentiles are used to give you a sense of where the bulk of the values in a set fall.

```python
# 1st quartile, Q1, lower
print(np.percentile(x, 25))

# 2nd quartile, Q2, median
print(np.percentile(x, 50))

# 3rd quartile, Q3, upper
print(np.percentile(x, 75))
```

<i>The 50th percentile is the same as the Median. The median cuts the data set in half. Half the data is below the median and the other half is above the median.</i>

<b>Linear Regression</b>

The term regression is used when we try to find the relationship between variables.

In Machine Learning, and in statistical modeling, that relationship is used to predict the outcome of future events.

Linear regression uses the relationship between the data-points to draw a straight line through all them.

    This line can be used to predict future values.

```python
# This is okay for predictions
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show() 
```

<b>R for Relationship</b>

It is important to know how the relationship between the values of the x-axis and the values of the y-axis is, if there are no relationship the linear regression can not be used to predict anything.

    This relationship - the coefficient of correlation - is called r.

The r value ranges from -1 to 1, where 0 means no relationship, and 1 (and -1) means 100% related.

Python and the Scipy module will compute this value for you, all you have to do is feed it with the x and y values.

```python
print(r)
```

The result -0.76 shows that there is a relationship, not perfect, but it indicates that we could use linear regression in future predictions.

```python
# predict speed of 10 years old cat
speed = myfunc(10)
print(speed)
```

The example predicted a speed at 85.6, which we also could read from the diagram. (where 10 intersects the line)

```python
# This data is not good for predictions
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show() 
```

```python
# should be a very low r value
slope, intercept, r, p, std_err = stats.linregress(x, y)
print(r)
```

The result: 0.013 indicates a very bad relationship, and tells us that this data set is not suitable for linear regression.

<b>Polynomial Regression</b>

If your data points clearly will not fit a linear regression (a straight line through all data points), it might be ideal for polynomial regression.

Polynomial regression, like linear regression, uses the relationship between the variables x and y to find the best way to draw a line through the data points.

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show() 
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show() 
```

```python
from sklearn.metrics import r2_score

mymodel = np.poly1d(np.polyfit(x, y, 3))
print(r2_score(y, mymodel(x))) 
```

    The result 0.94 shows that there is a very good relationship, and we can use polynomial regression in future predictions.

```python
mymodel = np.poly1d(np.polyfit(x, y, 3))

speed = mymodel(17)
print(speed) 
```

```python
# Bad fit
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(2, 95, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show() 
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 3))
print(r2_score(y, mymodel(x)))
```

    0.00995 indicates a very bad relationship, and tells us that this data set is not suitable for polynomial regression.

<b>Multiple Regression</b>

Multiple regression is like linear regression, but with more than one independent value, meaning that we try to predict a value based on two or more variables.

We can predict the CO2 emission of a car based on the size of the engine, but with multiple regression we can throw in more variables, like the weight of the car, to make the prediction more accurate.

    It is common to name the list of independent values with a upper case X, and the list of dependent values with a lower case y.

```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predictedCO2 = regr.predict([[2300, 1300]])

print(predictedCO2) 
```
